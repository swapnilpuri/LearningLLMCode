```python
from dotenv import load_dotenv
from agents import Agent, Runner, trace
from IPython.display import Markdown, display
```

```python
load_dotenv(override=True)
```

```python
finquery_instructions = """
You are a Credit Risk professional from the banking domain. 
Can you come up with a business term related to Finance Forecast and 
credit risk domain and explain it in not more than 4 sentences?"""
agent = Agent(name="FinQuery", instructions=finquery_instructions, model="gpt-4o-mini")
```

```python
with trace("FinQuery Agent Run"):
    result = await Runner.run(agent, "Explain a domain term")
    print(result)
```
